#### ***Origin txt***

```python
def get_list_of_files():
    
    files = glob.glob(os.path.join('/data/Public Data/IEEE Data Fusion Contest 2019/DFC2019_track4_trainval/Train-Track4/',"*_PC3.txt"))
    return sorted(files)

files=get_list_of_files()
```

```python
len(files)
```

#### Binary

```python
for i in range(len(files)):
    origintxtdata=np.array(pd.read_csv(files[i],sep=',',header=None))
    labelfile=files[i][:-7]+'CLS.txt'
    origintxtlabel=np.array(pd.read_csv(labelfile,header=None))
    print(origintxtdata.shape,origintxtlabel.shape)
    
    new_binlabel=origintxtlabel.copy()
    new_binlabel[origintxtlabel[:,0]==0,:]=0
    new_binlabel[origintxtlabel[:,0]==2,:]=1
    new_binlabel[origintxtlabel[:,0]==5,:]=2
    new_binlabel[origintxtlabel[:,0]==6,:]=2
    new_binlabel[origintxtlabel[:,0]==9,:]=2
    new_binlabel[origintxtlabel[:,0]==17,:]=2

    print('finishd numbers:{}'.format(i),new_binlabel.shape)
    
    ###Output
    path,datafilename=os.path.split(files[i])
    outputdatapath='/data/DW/DataFusion/dfc2019-master/track4/trn_data_folder/two_stage/stage1/Train-TXT/'+datafilename
    
    dt_data=pd.DataFrame(origintxtdata)
    dt_data.to_csv(outputdatapath,sep=',',header=None,index=False)
    
    path,labelfilename=os.path.split(labelfile)
    outputlabelpath='/data/DW/DataFusion/dfc2019-master/track4/trn_data_folder/two_stage/stage1/Train-TXT/'+labelfilename
    
    dt_label=pd.DataFrame(new_binlabel)
    dt_label.to_csv(outputlabelpath,sep=',',header=None,index=False)
    
```

#### Non_ground

```python
for i in range(len(files)):
    origintxtdata=np.array(pd.read_csv(files[i],sep=',',header=None))
    labelfile=files[i][:-7]+'CLS.txt'
    origintxtlabel=np.array(pd.read_csv(labelfile,header=None))
    print(origintxtdata.shape,origintxtlabel.shape)
    
    new_nonglabel=origintxtlabel[origintxtlabel[:,0]>2,:]
    new_nongdata=origintxtdata[origintxtlabel[:,0]>2,:]

    print('finishd numbers:{}'.format(i),new_nongdata.shape,new_nonglabel.shape)
    
    ###Output
    path,datafilename=os.path.split(files[i])
    outputdatapath='/data/DW/DataFusion/dfc2019-master/track4/trn_data_folder/two_stage/stage2/Train-TXT/'+datafilename
    
    dt_data=pd.DataFrame(new_nongdata)
    dt_data.to_csv(outputdatapath,sep=',',header=None,index=False)
    
    path,labelfilename=os.path.split(labelfile)
    outputlabelpath='/data/DW/DataFusion/dfc2019-master/track4/trn_data_folder/two_stage/stage2/Train-TXT/'+labelfilename
    
    dt_label=pd.DataFrame(new_nonglabel)
    dt_label.to_csv(outputlabelpath,sep=',',header=None,index=False)
    
```
